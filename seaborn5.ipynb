# Heat Map

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
import seaborn as sns
import pandas as pd
import numpy as np

```

```python
var=np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
plt.show()
```

```python
y1=sns.load_dataset('anagrams').head(10)


x=y1.drop(columns='attnr',axis=1)
x
```

```python
y1=sns.load_dataset('anagrams').head(10)


x=y1.drop(columns='attnr',axis=1)

sns.heatmap(x)
plt.show()
```

```python
y1=sns.load_dataset('anagrams').head(10)


x=y1.drop(columns='attnr',axis=1)

sns.heatmap(x,vmin=0,vmax=12) #to change color range to maximum and minimum use vmin=0,vmax=12
plt.show()
```

```python
y1=sns.load_dataset('anagrams').head(10)


x=y1.drop(columns='attnr',axis=1)

sns.heatmap(x,vmin=0,vmax=12,cmap='BrBG',annot=True) #to change color use cmap #annot to show values  on chart only
plt.show()
```

```python
var1=np.linspace(1,10,10).reshape(2,5)
var1


sns.heatmap(var1,vmin=0,vmax=12,cmap='BrBG',annot=True) 
plt.show()
```

```python
ar=np.array([['a0','a1','a2','a3','a4'],
             ['b0','b1','b2','b3','b4',]])
ar

sns.heatmap(var1,vmin=0,vmax=12,cmap='BrBG',annot=ar,fmt='s')  #to change the name of data
plt.show()
```

```python
var1=np.linspace(1,10,10).reshape(2,5)
var1

y={'fontsize':20,'color':'r'}
sns.heatmap(var1,vmin=0,vmax=12,cmap='BrBG',annot=True,annot_kws=y) #to change font size, color use annot_kws
plt.show()
```

```python
var1=np.linspace(1,10,10).reshape(2,5)
var1

y={'fontsize':10,'color':'r'}
sns.heatmap(var1,vmin=0,vmax=12,cmap='BrBG',
            annot=True,annot_kws=y,linewidth=4) #to give gap between bars
plt.show()
```

```python
var1=np.linspace(1,10,10).reshape(2,5)
var1

y={'fontsize':10,'color':'r'}
sns.heatmap(var1,vmin=0,vmax=12,cmap='BrBG',
            annot=True,annot_kws=y,linewidth=4,linecolor='y') #to change color of gap use linecolor
plt.show()
```

```python
var1=np.linspace(1,10,10).reshape(2,5)
var1

y={'fontsize':10,'color':'r'}
sns.heatmap(var1,vmin=0,vmax=12,cmap='BrBG',
            annot=True,annot_kws=y,linewidth=4,linecolor='y',cbar=False) #to remove colorbar use cbar=False else use cbar=True(which is bydefault)
plt.show()
```

```python
var1=np.linspace(1,10,10).reshape(2,5)
var1

y={'fontsize':10,'color':'r'}
sns.heatmap(var1,vmin=0,vmax=12,cmap='BrBG',
            annot=True,annot_kws=y,
            linewidth=4,linecolor='y',xticklabels=False,yticklabels=False) #to remove x and y axis labels
plt.show()
```

```python
var1=np.linspace(1,10,10).reshape(2,5)
var1

y={'fontsize':10,'color':'r'}
v= sns.heatmap(var1,vmin=0,vmax=12,cmap='BrBG',
            annot=True,annot_kws=y,
            linewidth=4,linecolor='y',xticklabels=False,yticklabels=False) #to remove x and y axis labels

v.set(xlabel="python",ylabel='No') #to give x and y label o0000000rrrrrrrr  #plt.xlabel("python",fontsize=15)

plt.show()
```
