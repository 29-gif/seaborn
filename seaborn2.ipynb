```python

y1=sns.load_dataset('penguins')

sns.barplot(x='island',y='bill_length_mm',data=y1,
            hue='sex') #hue will change graph according to sex #it will saperate combile sex graph to female and male

plt.show() 
```

```python

y1=sns.load_dataset('penguins')

order1 =['Dream','Biscoe','Torgersen']

sns.barplot(x='island',y='bill_length_mm',data=y1,
            hue='sex',order=order1,hue_order=['Female','Male']) #order=order1 to change order of bar name (dream,torgersen)#hue_order=['Female','Male'] to change bars of female and male ,which one you want to appear first

plt.show() 
```

```python

y1=sns.load_dataset('penguins')

order1 =['Dream','Biscoe','Torgersen']

sns.barplot(x='island',y='bill_length_mm',data=y1,
            hue='sex',order=order1,hue_order=['Female','Male']
            ,ci=100) #ci use to make line bigger at top of bar #ci should be between 0 to 100

plt.show() 
```

```python

y1=sns.load_dataset('penguins')

order1 =['Dream','Biscoe','Torgersen']

sns.barplot(x='island',y='bill_length_mm',data=y1,
            hue='sex',order=order1,hue_order=['Female','Male']
            ,ci=100,n_boot=5) #n_boot=5 is as same as ci
plt.show() 
```

```python

y1=sns.load_dataset('penguins')

order1 =['Dream','Biscoe','Torgersen']

sns.barplot(x='island',y='bill_length_mm',data=y1,
            hue='sex',order=order1,hue_order=['Female','Male']
            ,color='b') #to change color
plt.show() 
```

```python

y1=sns.load_dataset('penguins')

order1 =['Dream','Biscoe','Torgersen']

sns.barplot(x='island',y='bill_length_mm',data=y1,
            hue='sex',order=order1,hue_order=['Female','Male']
            ,palette='BrBG') #to change color palette='BrBG'
plt.show() 
```

```python

y1=sns.load_dataset('penguins')

order1 =['Dream','Biscoe','Torgersen']

sns.barplot(x='island',y='bill_length_mm',data=y1,
            hue='sex',order=order1,hue_order=['Female','Male']
            ,palette='BrBG',saturation=0.3) #to make graph light use saturation
plt.show() 
```

```python

y1=sns.load_dataset('penguins')

order1 =['Dream','Biscoe','Torgersen']

sns.barplot(x='island',y='bill_length_mm',data=y1,
            hue='sex',order=order1,hue_order=['Female','Male'],palette='BrBG',
            errcolor='y',errwidth=10) #to change color of error(line above bars) #use to change width of error (errwidth)
plt.show() 
```

```python

y1=sns.load_dataset('penguins')

order1 =['Dream','Biscoe','Torgersen']

sns.barplot(x='island',y='bill_length_mm',data=y1,
            hue='sex',order=order1,hue_order=['Female','Male'],palette='BrBG',
            errcolor='b',capsize=0.3) #cap will show end line
plt.show() 
```

```python

y1=sns.load_dataset('penguins')

order1 =['Dream','Biscoe','Torgersen']

sns.barplot(x='island',y='bill_length_mm',data=y1,
            hue='sex',order=order1,hue_order=['Female','Male'],palette='BrBG',
            errcolor='b',dodge=False)  #dodge=False it will make superimpose to bars
plt.show() 
```

```python
sns.set(style='darkgrid') #to make grid behind graphs   oooooooorrrrrrrr #plt.legend() #style='dark|darkgrid'

y1=sns.load_dataset('penguins')
order1 =['Dream','Biscoe','Torgersen']

sns.barplot(x='island',y='bill_length_mm',data=y1,
            hue='sex',order=order1,hue_order=['Female','Male'],palette='BrBG',
            errcolor='b',saturation=100)  #to blur the graph use alpha=0.5


plt.show() 
```
