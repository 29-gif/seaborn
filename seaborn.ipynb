# Line Plot

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
import seaborn as sns
import pandas as pd
```

```python
var =[1,2,3,4,5,6,7]
var1=[2,3,4,1,5,6,7]

'''
plt.plot(var,var1)
plt.show() #matplotlib
'''

x=pd.DataFrame({'var':var,'var1':var1})
sns.lineplot(x='var',y='var1',data=x)
```

```python
var =[1,2,3,4,5,6,7]
var1=[2,3,4,1,5,6,7]

'''
plt.plot(var,var1)
plt.show() #matplotlib
'''

x=pd.DataFrame({'var':var,'var1':var1})
sns.lineplot(x='var',y='var1',data=x)

plt.show() #use this to remove <Axes: xlabel='var', ylabel='var1'>
```

```python
#will give same output if not using dataframe but will remove default x and y axis labels
var =[1,2,3,4,5,6,7]
var1=[2,3,4,1,5,6,7]

sns.lineplot(x=var,y=var1)

plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(20)

sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=y1,hue='sex') #hue to use populer data of dataset

plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(20)

sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=y1,hue='sex',size=10)#to see size

plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(20)

sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=y1,hue='sex',style='sex') #to change style of any of the line

plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(20)

sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=y1,hue='sex',style='sex',palette="BrBG") #palette='Accent' to choose color

plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(20)

sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=y1,hue='sex',style='sex',palette="BrBG",markers=['o','>']) #markers=['o','>'] to give marker or point to graph

plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(20)

sns.lineplot(x='bill_length_mm',y='flipper_length_mm',
             data=y1,hue='sex',style='sex',palette="BrBG",
             markers=['o','>'],dashes=False) #dashes=False will remove ---- lines

plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(20)

sns.lineplot(x='bill_length_mm',y='flipper_length_mm',
             data=y1,hue='sex',style='sex',palette="BrBG",
             markers=['o','>'],dashes=False,legend='brief') #legend=False will remove legend #`legend` must be 'auto', 'brief', 'full', or a boolean.

plt.show() 
```

```python
sns.set(style='darkgrid') #to make grid behind graphs   oooooooorrrrrrrr #plt.legend() #style='dark|darkgrid'

plt.figure(figsize=(12,5))#to change figure size

sns.set_context("poster")#can use poster|paper| #to change text  #can use font_scale=2 to increse font size 

y1=sns.load_dataset('penguins').head(20)

sns.lineplot(x='bill_length_mm',y='flipper_length_mm',
             data=y1,hue='sex',style='sex',palette="BrBG",
             markers=['o','>'],dashes=False,legend='brief')

plt.title('Python')
sns.despine()#it will remove side line of graph
plt.show() 
```
