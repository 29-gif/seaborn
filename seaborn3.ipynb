# Scatter Plot

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
import seaborn as sns
import pandas as pd
```

```python
y1=sns.load_dataset('penguins')


y1
```

```python
y1=sns.load_dataset('penguins').head(30)

sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=y1)

plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(30)

sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=y1,hue='sex')

plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(30)

sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=y1,hue='sex',style='sex') # style='sex' to change style of data

plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(30)

sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',
                data=y1,hue='sex',style='sex',size='sex') # to change the size of scatters of female and male

plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(30)

sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',
                data=y1,hue='sex',style='sex',
                size='sex',sizes=(100,40)) #also use for sizes=(100,40) only but it will help to adjust the size according to your preferences 

plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(30)

sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',
                data=y1,hue='sex',style='sex',
                size='sex',sizes=(100,40)
               ,palette='gist_ncar') #to change color 
plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(30)

sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',
                data=y1,hue='sex',style='sex',
                size='sex',sizes=(100,40)
               ,palette='gist_ncar',alpha=0.2) #to make graph blur
plt.show() 
```

```python
y1=sns.load_dataset('penguins').head(30)
m={'Male':'o','Female':'*'}

sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',
                data=y1,hue='sex',style='sex',
                size='sex',sizes=(100,40)
               ,palette='gist_ncar',markers=m) #to changer marker
plt.show() 
```
