# Pair Plot

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
import seaborn as sns
import pandas as pd
```

```python
y1=sns.load_dataset('tips')
y1
```

```python
y1=sns.load_dataset('tips')
sns.pairplot(y1)#it will make pairplot for all numerical values and combination on y and x axis
plt.show()
```

```python
y1=sns.load_dataset('tips')
sns.pairplot(y1,hue='sex')
plt.show()
```

```python
y1=sns.load_dataset('tips')
sns.pairplot(y1,hue='sex',vars=['total_bill','tip'])# vars=['total_bill','tip'] to show only particular graphs
plt.show()
```

```python
y1=sns.load_dataset('tips')
sns.pairplot(y1,hue='sex',vars=['total_bill','tip'],
             hue_order=['Female','Male'])#to change order of female and order use hue_order
plt.show()
```

```python
y1=sns.load_dataset('tips')
sns.pairplot(y1,hue='sex',vars=['total_bill','tip'],
             hue_order=['Female','Male'],palette='BrBG')#to change order of female and order use hue_order
plt.show()
```

```python
y1=sns.load_dataset('tips')
sns.pairplot(y1,hue='sex',x_vars=['total_bill','tip'],
             hue_order=['Female','Male'],palette='BrBG')#x_vars=['total_bill','tip'] if you want only particular variables on x axis and y axis should remained same
```

```python
y1=sns.load_dataset('tips')
sns.pairplot(y1,hue='sex',y_vars=['total_bill','tip'],
             hue_order=['Female','Male'],
             palette='BrBG')#y_vars=['total_bill','tip'] if you want only particular variables on y axis and x axis should remained same
```

```python
y1=sns.load_dataset('tips')
sns.pairplot(y1,hue='sex',hue_order=['Female','Male']
             ,kind='reg',palette='BrBG') #to change kind of graph #kind='reg|scatter|kde|hist' #by default its scatter
plt.show()
```

```python
#change kind
y1=sns.load_dataset('tips')
sns.pairplot(y1,hue='sex',hue_order=['Female','Male'],kind='kde',palette='BrBG') #to change kind of graph #kind='reg|scatter|kde|hist'
plt.show()
```

```python
#change kind
y1=sns.load_dataset('tips')
sns.pairplot(y1,hue='sex',hue_order=['Female','Male'],kind='hist',palette='BrBG') #to change kind of graph #kind='reg|scatter|kde|hist'
plt.show()
```

```python

y1=sns.load_dataset('tips')

sns.pairplot(y1,hue='sex',hue_order=['Female','Male'],kind='kde',
             palette='BrBG',diag_kind='hist') #to change kind of graph diagonaly to hist using diag_kind='hist' 
plt.show()
```

```python

y1=sns.load_dataset('tips').head(10)

sns.pairplot(y1,hue='sex',hue_order=['Female','Male'],markers=['*','<']) #to change markers
plt.show()
```
