# CountPlot

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
import seaborn as sns
import pandas as pd
```

```python
y1=sns.load_dataset('tips')


y1
```

```python
y1=sns.load_dataset('tips')
sns.countplot(x='sex',data=y1)

plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.countplot(x='sex',data=y1,hue='smoker') 

plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.countplot(y='sex',data=y1,hue='smoker') #to make graph horizontal just change x data to y

plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.countplot(x='sex',data=y1,hue='smoker',
              palette='bwr',saturation=0.3)  #to make graph color light use saturation


plt.show() 
```
