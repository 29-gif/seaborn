# Multiple Plots |Facet-Grid

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
import seaborn as sns
import pandas as pd
```

```python
y1=sns.load_dataset('tips').head(20)


y1
```

```python
y1=sns.load_dataset('tips')

fg=sns.FacetGrid(y1)

fg.map(plt.scatter,'total_bill','tip')#can use any graph instead of scatter

plt.show() 
```

```python
y1=sns.load_dataset('tips')

fg=sns.FacetGrid(y1,col='sex')#to make multiple graph

fg.map(plt.scatter,'total_bill','tip')

plt.show() 
```

```python
#day wise alg kryu
y1=sns.load_dataset('tips')

fg=sns.FacetGrid(y1,col='sex',hue='day')

fg.map(plt.scatter,'total_bill','tip').add_legend() 


plt.show() 
```

```python
y1=sns.load_dataset('tips')

fg=sns.FacetGrid(y1,col='day',hue='sex') #same graph just changed day to sex and sex to day

fg.map(plt.bar,'total_bill','tip').add_legend() 


plt.show() 
```

```python
y1=sns.load_dataset('tips')

fg=sns.FacetGrid(y1,col='day',hue='sex') #same graph just changed day to sex and sex to day

fg.map(plt.bar,'total_bill','tip',edgecolor='g').add_legend() 


plt.show() 
```
