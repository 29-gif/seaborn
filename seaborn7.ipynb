# Violin Plot

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
import seaborn as sns
import pandas as pd
```

```python
y1=sns.load_dataset('tips')
y1
```

```python
y1=sns.load_dataset('tips')
sns.violinplot(x='day',y='total_bill',data=y1)
plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.violinplot(x='day',y='total_bill',data=y1,hue='time')

plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.violinplot(x='day',y='total_bill',data=y1,palette='BrBG',linewidth=5) #to give a line

plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.violinplot(x='time',y='total_bill',data=y1,palette='BrBG',order=['Dinner','Lunch'])  #to change order of x labels

plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.violinplot(x='time',y='total_bill',data=y1,palette='BrBG',order=['Dinner','Lunch'],saturation=0.2) 

plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.violinplot(x='day',y='total_bill',data=y1,palette='BrBG',hue='sex',split=True) #to split male and female data

plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.violinplot(x='day',y='total_bill',data=y1,palette='BrBG',
               hue='sex',split=True,scale='count') #scale can be (count|area|with)
plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.violinplot(x='total_bill',y='day',data=y1,palette='BrBG',
               hue='sex') #to make graph horizontally just change x and y axis
plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.violinplot(x='time',y='total_bill',data=y1,palette='BrBG',
               order=['Dinner','Lunch'],inner='quart') #inner can be(quart|point|stick|none)

plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.violinplot(x='time',y='total_bill',data=y1,palette='BrBG',
               order=['Dinner','Lunch'],inner='point') #inner can be(quart|point|stick|none)

plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.violinplot(x=y1['total_bill']) #to make single violin plot

plt.show() 
```

```python
y1=sns.load_dataset('tips')
sns.violinplot(y=y1['total_bill']) #to make violin vertical

plt.show() 
```
